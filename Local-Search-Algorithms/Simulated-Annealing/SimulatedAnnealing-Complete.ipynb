### Making the simulated annealing demonstration

```python
import math
import random
import matplotlib.pyplot as plt


def fcost(sol):
    return sum([i**2 for i in sol])


def successors(sol, step=1.0):
    succ = [x + random.uniform(-step, step) for x in sol]
    return succ


def simAnnealing(initSol, initTemp, alpha, iters):
    currSol = initSol
    cost = fcost(currSol)
    sol = currSol
    minCost = cost
    temp = initTemp

    for iteration in range(iters):
        neighbor = successors(currSol)
        ncost = fcost(neighbor)

        costdiff = ncost - cost

        if costdiff < 0 or random.random() < math.exp(-costdiff / temp):
            currSol = neighbor
            cost = ncost

            if cost < minCost:
                sol = currSol
                minCost = cost

        temp *= alpha

    return sol, minCost


initSol = [300.0, 400.0]
initTemp = 1000.0
alpha = 0.95
iters = 500
x= []
y= []
for i in range(iters):
    bestSol, cost = simAnnealing(
    initSol, initTemp, alpha, iters)
    x.append(bestSol[0])
    y.append(bestSol[1])

plt.scatter(x,y)
plt.scatter(300,400,c="red")  
print("\nBest Solution:", bestSol)
print("Best Cost:", cost)
plt.scatter(bestSol[0],bestSol[1],c="yellow")
```
